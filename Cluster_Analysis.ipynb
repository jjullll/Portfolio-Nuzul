```python
# Mepersiapkan library yang akan digunakan
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Pembuatan toy data mengunakan make blobs
# cluster std untuk mengatur jarak data
# random state untuk menyeragamkan data disetiap centroid
# y true = untuk mengkategorikan data setiap barisnya
x, y_true = make_blobs(n_samples=300,
                       n_features=2,
                       centers=4,
                       cluster_std=1.3,
                       random_state=13)
```

```python
x
```

```python
# s adalah ukuran dari lingkaran data
plt.scatter(x[:,0],x[:,1],s=50)
```

```python
kmeans = KMeans(n_clusters=4)

kmeans.fit(x)
```

```python
y_kmeans = kmeans.predict(x)
```

```python
# c digunakan untuk menambah parameter warna
plt.scatter(x[:,0],x[:,1],s=50,c=y_kmeans)

# menambahkan centeroid pada clustering yang telah dibuat
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1],c='red')
```
